```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import pickle
```

```python
# Memanggil Data yang akan digunakan
data = pd.read_csv('water_potability.csv')
data
```

```python
# Mengisi data yang kosong dengan nilai rata-rata setiap kolom
data = data.fillna(data.mean())
```

```python
data['Potability'] = data['Potability'].apply(lambda x: 'Layak minum' if x == 1 else ('Tidak Layak  Minum' if x ==0 else None))
data
```

```python
# Memisahkan data Label dan data fitur
data_label = data["Potability"]
data_fitur = data[data.columns[0:9]]
```

```python
# standarisasi nilai-nilai dari dataset
scaler = StandardScaler()
scaler.fit(data_fitur)
data_fitur = scaler.transform(data_fitur)
```

```python
# Membagi data menjadi data latih dan data uji
X_train, X_test, Y_train, Y_test = train_test_split(data_fitur, data_label, test_size = 0.2, random_state=42)

# Membagi data menjadi data latih dan data uji, lalu data latih dibagi menjadi data val untuk dipakai pada model neural network
X_truetrain, X_val, Y_truetrain, Y_val = train_test_split(X_train, Y_train, test_size = 0.3, random_state=42)
```

```python
# Melatih Model dengan parameter yang telah didatapatkan
classifier = MLPClassifier(hidden_layer_sizes=(200, ), activation='tanh')
classifier.fit(X_truetrain, Y_truetrain)
```

```python
pickle.dump(classifier, open("model1.pkl","wb"))
```
